##**Systemy rekomendacji**


System rekomendacji śledzi i analizuje wsadowe dane, proponując odwiedzającym nowe treści lub produkty.

Dzięki wykorzystaniu technologii sztucznej inteligencji i maszynowego uczenia, silnik jest w stanie dobrać rekomendacje w taki sposób, że z dużym prawdopodobieństwem matematycznym spodobają się konkretnemu użytkownikowi.

Obecnie silniki rekomendacji mają szerokie zastosowanie w dużych serwisach internetowych takich jak: wypożyczalnie filmów online, serwisy streamingowe, serwisy informacyjne, księgarnie internetowe oraz w wielu innych miejscach w sieci.

Wartym uwagi przykładem skutecznego zastosowania inteligentnych rekomendacji jest najbardziej popularna aplikacja 2020 roku - TikTok, gdzie został wdrożony jeden z najbardziej spersonalizowanych systemów rekomendacji w branży.

Dzięki temu rozwiązaniu użytkownicy nie muszą sami szukać tego, co obejrzeć. Algorytm TikToka ich w tym wyręcza, dopasowując się do ich upodobań i nawyków. TikTok może się pochwalić jednym z najwyższych współczynników zaangażowania wśród użytkowników w porównaniu do Facebooka czy Instagrama.



**Systemy rekomendacji dzielimy na trzy grupy:**
1.   **Content-Based Filtering** - metoda filtrowania treści oparta na zbieraniu i analizowaniu informacji oraz danych użytkowników na temat ich zachowania w sklepie.
Następnie tworzy listę rekomendowanych produktów na podstawie historii przeglądania danego użytkownika.
Założeniem filtrowania opartego na treści jest to, że jeśli klientowi spodoba się określony produkt, to istnieje prawdopodobieństwo, że spodoba mu się również inny produkt o podobnej specyfice, o podobnym opisie. Przykład:  rekomendacje wideo na YouTube’ie.








Content-Based Filtering.jpg

2. **Collaborative Filtering** - analizuje dane uzyskane od użytkowników, którzy kupili podobne produkty, a następnie łączy te informacje w celu utworzenia listy polecanych produktów dla kolejnego użytkownika.
Dużą zaletą tej metody jest to, że umożliwia silnikowi rekomendacji generowanie propozycji dla stosunkowo skomplikowanych produktów takich jak filmy czy muzyka bez konieczności posiadania obszernej wiedzy na ich temat.
Ta metoda oparta jest na założeniu, że klienci będą preferować rekomendacje oparte na wyborach, które dokonali inni użytkownicy w przeszłości.
Przykład: Jeśli klient A przeglądał T-shirty marki X oraz marki Y, to klientowi B, który również zaczął oglądać te produkty, będą wyświetlone rekomendacje wygenerowane na podstawie historii przeglądania klienta A.

Collaborative filtering.jpg

3. **Hybrydowa metoda filtrowania** - System rekomendacji oparty na metodzie hybrydowej łączy systemy oparte na treści oraz te oparte na grupowym filtrowaniu.
W tym przypadku silnik rekomendacji filtruje zarówno informacje o zachowaniu innych użytkowników z przeszłości, jak i preferencjach danego klienta, dla którego rekomendacje są wyświetlane.
Tym samym metoda ta łączy najlepsze aspekty dwóch poprzednich w jedno, spójne rozwiązanie.
Przykład: Przykładem hybrydowego systemu rekomendacji może być sposób, w jaki Spotify dobiera utwory do spersonalizowanych playlist “Discover Weekly”. Użytkownicy często są pod wrażeniem tego, jak wybrane przez algorytm piosenki trafiają w ich gust muzyczny.

##One-hot encoding

Istotą one-hot encoding jest zastąpienie kategorii zmiennej poprzez ciąg binarny o długości n z jedną „1” w ciągu. Modyfikacją tego kodowania jest wykorzystanie ciągu binarnego o długości n-1, z uwagi na to że ciąg binarny o takiej długości zapewnia pełną informację o kategoriach zmiennej. Kodowanie poprzez ciąg „n-1” określony jest jako kodowanie ze ślepą zmienną (ang. Dummy Variable Encoding) i jest powszechnie zalecany przy wdrożeniach modeli uczenia maszynowego. Jednakże są przypadki, gdy n-elementowy ciąg może być zasadniejszy do użycia. Do nich należy choćby sytuacja, gdy interesuje nas wpływ pojedynczej kategorii lub budujemy modele oparte na drzewach decyzyjnych. Kodowanie One-Hot Encoding oprócz swoich zalet ma też i wady, do których należy rozszerzenie przestrzeni zmiennych przy braku dodania dodatkowych informacji. W przypadku wielu zmiennych kategorialnych czy nawet dużej ilości kategorii w zmiennej ma to znaczenie.

```python
#sprawdzenie jakie kolumny są w df - pomoc przy dalszym tworzeniu kodu
dog_list.columns
```

# 4.Modelowanie


##Wstępne przetwarzanie danych

Na etapie modelowania wykorzystano funkcję prawdopodobieństwa - podobieństwo kosinusowe. Jako jeden z możliwych systemów rekomendacyjnych zastosowano Colaborative Filtering, czyli system badający odległości na bazie macierzy cech.

W celu odpowiedniego wyszukania danych cech, połączono wszystkie kolumny, które będą miały wpływ na wyszukanie odpowiedniego zwierzaka w utworzonych systemie rekomendacyjnym. Utworzono zmienną o nazwie **combined_features** i w niej przechowujemy połączenie wszystkich kolumn.

```python
#łączenie wszystkich 5 wybranych cech
combined_features = dog_list['Height']+' '+dog_list['Weight']+' '+dog_list['Shedding_category']+' '+dog_list['Energy_level_category']+' '+dog_list['Demeanor_category']
print(combined_features)
```

Kolejnym krokiem jest przekonwertowanie danych tekstowych na wektory cech - wartości liczbowe.  Jedynie wartości podobieństwa dobierane są w oparciu o przekształcenie ich na wektor.     

```python
#przekształcanie danych tekstowych na wektory cech
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

Otrzymaliśmy połączone cechy i zostały one dopasowane oraz przekształcone na wartości numeryczne, aby w dalszym etapie tworzenia systemu rekomendacyjnego, łatwiej było odnosić się do danych wartości i znaleźć rekordy (psy) o idealnej trafności.

Pierwszy indeks to dany pies, a drugi to wyniki podobieństwa - każdego psa, porównujemy z pozostałymi

##Podobieństwo cosinusowe

Następnym krokiem po przekształceniu z wartości tekstowych na numeryczne, jest znalezienie podobieństwa. Do tego celu użyjemy **podobieństwa kosinusowego**. W analizie danych podobieństwo kosinusowe jest miarą podobieństwa między dwoma niezerowymi wektorami zdefiniowanymi w wewnętrznej przestrzeni iloczynu. Podobieństwo kosinusów to cosinus kąta między wektorami (iloczyn skalarny wektorów podzielonych przez iloczyn ich długości).

Ma to na celu znalezienie wartości najbardziej do siebie pasujących, podobnych - jak najlepsze dopasowanie psa do wskazanych cech.

```python
#uzyskanie wyników podobieństwa przy użyciu podobieństwa cosinusowego
similarity = cosine_similarity(feature_vectors)
print(similarity)
```

Otrzymana macierz ma wymiary 1622 na 1622, co oznacza, że wszystkie wartości ze zbioru zostały uwzględnione.

Wynik sprowadzamy do listy w oparciu o indeks (Id). Pierwsza wartość to kolejne wiersze data frame - Id. Drugi natomiast to podobieństwo dopasowania danej wartości z psem, zwierzakiem o Id = 6.

W celu uporządkowania danych na liście, otrzymane wyniki przedstawiono w kolejności od najlepszego dopasowania do najgorszego. Widzimy, że jest sporo psów z dopasowaniem idealnym. Prawdopodobnie są to przedstawiciele tej samej rasy.

```python
#sortowanie psów na podstawie ich wyniku podobieństwa
sorted_similar_dogs = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_dogs)
```

Ostatni krok na tym etapie to sprecyzowanie jakie wartości chcemy, aby zostały wyświetlone oraz ile podobnych ras psów do siebie chcemy zaproponować przyszłym opiekunom.

```python
#wypisuje nazwy podobnych psów na podstawie indeksu
print('Psy dla Ciebie to: \n')

i = 1

for dog in sorted_similar_dogs:
  index = dog[0]
  dog_from_index = dog_list[dog_list.index==index]['Id'].values[0]
  dog_from_index2 = dog_list[dog_list.index==index]['Breed'].values[0]
  dog_from_index3 = dog_list[dog_list.index==index]['Location'].values[0]
  dog_from_index4 = dog_list[dog_list.index==index]['Destiny'].values[0]
  dog_from_index5 = dog_list[dog_list.index==index]['Height'].values[0]
  dog_from_index6 = dog_list[dog_list.index==index]['Weight'].values[0]
  dog_from_index7 = dog_list[dog_list.index==index]['Group'].values[0]
  dog_from_index8 = dog_list[dog_list.index==index]['Grooming_frequency_category'].values[0]
  dog_from_index9 = dog_list[dog_list.index==index]['Shedding_category'].values[0]
  dog_from_index10 = dog_list[dog_list.index==index]['Energy_level_category'].values[0]
  if (i<4) and  dog_from_index4 == 'Gotowy Do Adopcji':
    print(i, '.',dog_from_index, ',', dog_from_index2, ',', dog_from_index3, ',', dog_from_index5, ',', dog_from_index6, ',', dog_from_index7, ',',dog_from_index8, ',', dog_from_index9, ',',dog_from_index10)
    i+=1
```
